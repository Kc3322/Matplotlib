```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 2, 4, 6])
ypoints = np.array([0, 80, 120, 250])

plt.plot(xpoints, ypoints, '*')
plt.show()
```

```python

ypoints = np.array([0, 80, 120, 250])

plt.plot( ypoints, '*:k', ms=20)  #marker(*)+line(:)+color(k)   ms-marker size
plt.show()
```

```python
ypoints = np.array([4, 6, 4, 1])

plt.plot(ypoints, ls = '-.',linewidth = '3', marker='o', c = 'g')


plt.show()
```

```python
#multiple lines
y1points = np.array([4, 6, 3,5,7,3,8,2,8,1])
y2points = np.array([3, 0, 9,4,12,9,11,6,7,12])
plt.plot(y1points, ls = '--',linewidth = '3', marker='o', c = 'g')
plt.plot(y2points, ls='-',marker='*',c='r')


plt.show()
```

```python
x = np.array([2020, 2021,2022, 2023, 2024])
y = np.array([10000,  500000,  150000, 30000 , 40])

plt.plot(x, y,marker='o')

plt.title("COVID-19 DEATH RATE")
plt.xlabel("YEAR")
plt.ylabel("number of casualties")

plt.grid(ls='--',color='r')  #grid lines
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3, 1,1.3, 2, 3])
y = np.array([1, 7, 10, 11, 8])

plt.subplot(1, 2, 1)
plt.plot(x,y,marker="o",color="r")


x = np.array([1, 3, 2.7, 2, 1])
y = np.array([1, 20, 30,33, 23])

plt.subplot(1, 2, 2)
plt.plot(x,y,marker="o",color="r")

plt.show()
```
